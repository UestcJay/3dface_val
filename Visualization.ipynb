# Vizualize NME

```python
import os
import matplotlib
%matplotlib inline
import pylab as plt
import numpy as np

if not os.path.isdir('out/NME_viz'):
    os.makedirs('out/NME_viz')

datasets = ['Menpo-3D', '300W-Testset-3D', 'AFLW2000-3D-Reannotated',
            '300VW-3D-CatA', '300VW-3D-CatB', '300VW-3D-CatC']
for dataset in datasets:
    errors_proposed = np.load('results/HGFPN/%s.npy' % dataset)
    errors_3ddfa = np.load('results/3DDFA/%s.npy' % dataset)
    errors_3dfan = np.load('results/3D-FAN/without_plus/%s.npy' % dataset)
    axes1 = np.linspace(0, 1, 1000)
    axes2_proposed = np.zeros(1000)
    axes2_3dfan = np.zeros(1000)
    axes2_3ddfa = np.zeros(1000)
    f = plt.figure()
    for i in range(1000):
        axes2_proposed[i] = (errors_proposed < axes1[i]).sum() / float(errors_proposed.shape[0])
        axes2_3ddfa[i] = (errors_3ddfa < axes1[i]).sum() / float(errors_3ddfa.shape[0])
        axes2_3dfan[i] = (errors_3dfan < axes1[i]).sum() / float(errors_3dfan.shape[0])
    plt.xlim(0,7)
    plt.ylim(0,100)
    plt.yticks(np.arange(0,110,10))
    plt.xticks(np.arange(0,8,1))
    plt.grid()
    plt.title('NME (%)', fontsize=20)
    plt.xlabel('NME (%)', fontsize=16)
    plt.ylabel('Test Images (%)', fontsize=16)
    plt.plot(axes1*100,axes2_proposed*100,'r-',label='Proposed',lw=3)
    plt.plot(axes1*100,axes2_3dfan*100,'g-',label='3D-FAN',lw=3)
    plt.plot(axes1*100,axes2_3ddfa*100,'b-',label='3DDFA',lw=3)
    plt.legend(loc=4, fontsize=16)
    # plt.show()
    plt.savefig('out/NME_viz/' + dataset + '.pdf', transparent=True, bbox_inches='tight', pad_inches=0)
```
